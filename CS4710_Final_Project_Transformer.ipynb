```python
import pretty_midi
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader

def midi_to_sequence(midi):
    """
    Convert MIDI to a sequence of note events.
    """
    notes = []
    for instrument in midi.instruments:
        if not instrument.is_drum:
            for note in instrument.notes:
                notes.append((note.start, note.pitch, note.velocity, note.end - note.start))
    notes.sort(key=lambda x: x[0])  # Sort by start time
    return notes

def sequence_to_tensor(sequence, max_length=512):
    """
    Convert a sequence of notes to a tensor.
    """
    tensor = []
    for start, pitch, velocity, duration in sequence:
        tensor.append([start, pitch, velocity, duration])
    tensor = torch.tensor(tensor, dtype=torch.float32)
    if len(tensor) > max_length:
        tensor = tensor[:max_length]  # Truncate
    else:
        # Pad with zeros
        padding = torch.zeros((max_length - len(tensor), 4), dtype=torch.float32)
        tensor = torch.cat([tensor, padding], dim=0)
    return tensor

class MidiDataset(Dataset):
    def __init__(self, midi_files, genre_labels, genre_to_index):
        self.midi_files = midi_files
        self.genre_indices = [genre_to_index[label] for label in genre_labels]

    def __len__(self):
        return len(self.midi_files)

    def __getitem__(self, idx):
        midi = self.midi_files[idx]
        genre_index = self.genre_indices[idx]
        sequence = midi_to_sequence(midi)
        tensor = sequence_to_tensor(sequence)
        genre_tensor = torch.tensor(genre_index, dtype=torch.long)
        return tensor, genre_tensor
```

```python
import torch.nn as nn

class MidiTransformer(nn.Module):
    def __init__(self, input_dim, genre_dim, embed_dim, num_heads, ff_dim, num_layers):
        super(MidiTransformer, self).__init__()
        self.embedding = nn.Linear(input_dim, embed_dim)
        self.genre_embedding = nn.Embedding(genre_dim, embed_dim)
        self.positional_encoding = nn.Parameter(torch.zeros(512, embed_dim))

        self.transformer = nn.Transformer(
            d_model=embed_dim,
            nhead=num_heads,
            num_encoder_layers=num_layers,
            num_decoder_layers=num_layers,
            dim_feedforward=ff_dim
        )

        self.output_layer = nn.Linear(embed_dim, input_dim)

    def forward(self, x, genre):
        # Add embeddings
        x_embed = self.embedding(x) + self.positional_encoding[:x.size(1)]
        genre_embed = self.genre_embedding(genre).unsqueeze(1)

        # Use genre embedding as the decoder input
        decoder_input = genre_embed.repeat(1, x.size(1), 1)
        output = self.transformer(x_embed, decoder_input)

        # Project back to MIDI token space
        output = self.output_layer(output)
        return output
```

```python
def train_model(model, dataloader, optimizer, criterion, epochs=10, device='cuda'):
    model.to(device)
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for inputs, genres in dataloader:
            inputs, genres = inputs.to(device), genres.to(device)
            optimizer.zero_grad()
            outputs = model(inputs, genres)
            loss = criterion(outputs, inputs)  # Reconstruction loss
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {total_loss / len(dataloader)}")
```

```python
def tensor_to_midi(tensor):
    """
    Convert a tensor back to a MIDI file.
    """
    midi = pretty_midi.PrettyMIDI()
    instrument = pretty_midi.Instrument(program=0)
    for event in tensor:
        start, pitch, velocity, duration = event
        if duration > 0:  # Ignore padding
            note = pretty_midi.Note(
                velocity=int(velocity),
                pitch=int(pitch),
                start=float(start),
                end=float(start + duration)
            )
            instrument.notes.append(note)
    midi.instruments.append(instrument)
    return midi

def generate_midi(model, input_midi, genre_index, device='cuda'):
    model.eval()
    input_tensor = sequence_to_tensor(midi_to_sequence(input_midi)).unsqueeze(0).to(device)
    genre_tensor = torch.tensor([genre_index], dtype=torch.long).to(device)
    with torch.no_grad():
        output_tensor = model(input_tensor, genre_tensor)
    return tensor_to_midi(output_tensor.squeeze(0).cpu().numpy())
```

```python
def add_duration_patterns(base_duration, genre, beat_position):
    """Add genre-specific rhythmic patterns"""
    genre_patterns = {
        'Jazz': {
            'swing_ratio': 1.2,
            'pattern': [1.2, 0.8, 1.1, 0.9]
        },
        'Electronic': {
            'swing_ratio': 1.0,
            'pattern': [1.0, 1.0, 1.0, 1.0]  # Strict quantization
        },
        'Rock': {
            'swing_ratio': 1.1,
            'pattern': [1.1, 0.9, 1.0, 0.9]
        },
        'Hip-Hop': {
            'swing_ratio': 1.15,
            'pattern': [1.2, 0.8, 1.1, 0.8]
        },
        'Latin': {
            'swing_ratio': 1.1,
            'pattern': [1.1, 0.9, 1.1, 0.9]
        },
        'Dance': {
            'swing_ratio': 1.0,
            'pattern': [1.0, 1.0, 1.0, 1.0]
        }
    }

    default_pattern = {
        'swing_ratio': 1.0,
        'pattern': [1.0, 1.0, 1.0, 1.0]
    }

    pattern_info = genre_patterns.get(genre, default_pattern)
    pattern = pattern_info['pattern']

    # Apply pattern based on beat position
    modified_duration = base_duration * pattern[beat_position % 4]

    # Quantize duration for electronic genres
    if genre in ['Electronic', 'Dance']:
        modified_duration = round(modified_duration * 4) / 4

    return modified_duration

def add_rhythmic_patterns(params, current_time, velocity):
    """Add genre-specific velocity accents"""
    genre = params.get('genre', 'default')

    accent_patterns = {
        'Rock': [1.2, 0.9, 1.1, 0.9],  # Emphasis on 1 and 3
        'Jazz': [1.1, 0.9, 1.0, 0.95],  # Subtle swing
        'Hip-Hop': [1.2, 0.8, 1.1, 0.9],  # Strong beats
        'Dance': [1.1, 1.0, 1.1, 1.0],  # Four on the floor
        'Latin': [1.1, 0.9, 1.1, 0.95],  # Latin clave feel
        'Electronic': [1.1, 1.0, 1.1, 1.0],  # Regular accents
        'Country': [1.1, 0.9, 1.0, 0.9],  # Traditional feel
        'Folk': [1.05, 0.95, 1.0, 0.95],  # Gentle accents
        'R&B': [1.1, 0.9, 1.05, 0.95],  # Smooth groove
        'Pop': [1.1, 0.95, 1.05, 0.95]  # Balanced accents
    }

    default_pattern = [1.0, 1.0, 1.0, 1.0]
    pattern = accent_patterns.get(genre, default_pattern)

    # Calculate beat position (assuming 4/4 time)
    beat_position = int((current_time * 4) % 4)

    # Apply accent
    modified_velocity = int(velocity * pattern[beat_position])
    return min(modified_velocity, 127)  # Ensure we don't exceed MIDI velocity limit
```

```python
def normalize_durations(notes, target_duration):
    """Scale durations while preserving rhythmic relationships"""
    if not notes:
        return notes

    current_duration = max(note.end for note in notes)
    if current_duration == 0:
        return notes

    scale_factor = target_duration / current_duration

    # Preserve relative note spacing
    for note in notes:
        note.start = note.start * scale_factor
        note.end = note.end * scale_factor

        # Ensure minimum duration
        if note.end - note.start < 0.1:
            note.end = note.start + 0.1

    return notes


def balance_note_density(notes, target_density, duration, min_pitch=0, max_pitch=127):
    """Adjust note density while preserving musical structure"""
    if not notes:
        return notes

    current_density = len(notes) / duration
    target_count = int(target_density * duration)

    if len(notes) < target_count * 0.7:  # Generate more notes
        additional_notes = []
        notes_needed = target_count - len(notes)

        for i in range(min(notes_needed, len(notes))):
            base_note = notes[i]
            for octave in [-12, 0, 12]:  # Add octave variations
                new_note = pretty_midi.Note(
                    velocity=base_note.velocity,
                    pitch=base_note.pitch + octave,
                    start=base_note.start + 0.05,
                    end=base_note.end + 0.05
                )
                if min_pitch <= new_note.pitch <= max_pitch:
                    additional_notes.append(new_note)

        return notes + additional_notes

    return notes
```

```python
def tensor_to_midi(tensor, genre, tempo=120.0):
   pm = pretty_midi.PrettyMIDI(initial_tempo=tempo)
   instrument = pretty_midi.Instrument(program=0)

   params = get_enhanced_genre_params().get(genre, DEFAULT_PARAMS)
   params['genre'] = genre

   # Adjust time scale based on genre
   params['time_scale'] = params['time_scale'] * 2.0

   # Get pitch range
   min_pitch, max_pitch = params['pitch_range']

   current_time = 0.0
   last_pitch = None
   pitch_direction = 1
   pitch_step_size = 0

   generated_notes = []

   for i in range(len(tensor)):
       time_delta, pitch_pred, velocity_pred, duration_pred = tensor[i]

       if np.random.random() < params['note_density']:
           # Enhanced pitch selection with genre constraints
           if last_pitch is None:
               pitch = int(np.clip(
                   np.interp(pitch_pred, [0, 1], [min_pitch, max_pitch]) +
                   np.random.normal(0, 2),
                   min_pitch,
                   max_pitch
               ))
           else:
               if np.random.random() < 0.2:
                   pitch_direction *= -1
                   pitch_step_size = np.random.randint(*params['pitch_step_range'])

               new_pitch = last_pitch + (pitch_direction * pitch_step_size)

               if new_pitch < min_pitch or new_pitch > max_pitch:
                   pitch_direction *= -1
                   new_pitch = last_pitch + (pitch_direction * pitch_step_size)

               pitch = int(np.clip(new_pitch, min_pitch, max_pitch))

           # Enhanced duration with musical phrasing
           base_duration = float(np.interp(duration_pred, [0, 1],
                              [params['min_duration'], params['max_duration']]))

           beat_position = int((current_time * 4) % 4)
           duration = add_duration_patterns(base_duration, genre, beat_position)

           # Longer notes with reduced probability
           if np.random.random() < 0.15:  # Increased from 0.1
               duration *= 1.5  # Reduced from 2.0

           base_velocity = int(np.interp(velocity_pred, [0, 1], params['velocity_range']))
           velocity = add_rhythmic_patterns(params, current_time, base_velocity)

           note = pretty_midi.Note(
               velocity=velocity,
               pitch=pitch,
               start=current_time,
               end=current_time + duration
           )

           generated_notes.append(note)
           last_pitch = pitch

       current_time += float(time_delta) * params['time_scale']

       if current_time > 240:  # Increased from 180s
           break

   # Calculate target duration and density
   target_duration = min(240.0, current_time)  # Increased max duration
   target_density = params.get('target_density', 20.0) * 1.2  # Increased density

   # Post-process notes
   generated_notes = normalize_durations(generated_notes, target_duration)
   generated_notes = balance_note_density(
       generated_notes,
       target_density,
       target_duration,
       min_pitch=min_pitch,
       max_pitch=max_pitch
   )

   instrument.notes = generated_notes
   pm.instruments.append(instrument)
   return pm
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import pretty_midi
import glob
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

class EnhancedMidiLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, genre_dim, num_layers, max_sequence_length=2048):
        super(EnhancedMidiLSTM, self).__init__()

        # Genre-specific tempo and density parameters
        self.genre_tempo_base = nn.Embedding(genre_dim, 1)
        self.genre_density_scale = nn.Embedding(genre_dim, 1)

        # Expanded pitch range handling
        self.pitch_range_low = nn.Embedding(genre_dim, 1)
        self.pitch_range_high = nn.Embedding(genre_dim, 1)

        # Initialize with genre-specific defaults
        self.genre_defaults = {
            'Electronic': {'tempo': 128, 'density': 1.2},
            'Dance': {'tempo': 130, 'density': 1.3},
            'Jazz': {'tempo': 100, 'density': 0.8},
            'Rock': {'tempo': 120, 'density': 1.0},
            'Hip-Hop': {'tempo': 95, 'density': 0.9},
            'Folk': {'tempo': 90, 'density': 0.7},
            'Latin': {'tempo': 110, 'density': 1.1},
            'R&B': {'tempo': 95, 'density': 0.9},
            'World': {'tempo': 100, 'density': 1.0},
            'Pop': {'tempo': 120, 'density': 1.0},
            'Country': {'tempo': 100, 'density': 0.8},
            'Reggae': {'tempo': 90, 'density': 0.7}
        }

        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.max_sequence_length = max_sequence_length

        # Enhanced input processing
        self.note_embedding = nn.Linear(input_dim, hidden_dim)
        self.genre_embedding = nn.Embedding(genre_dim, hidden_dim)

        # Position encoding for better sequence handling
        self.register_buffer(
            "position_ids",
            torch.arange(max_sequence_length).expand((1, -1))
        )
        self.position_embedding = nn.Embedding(max_sequence_length, hidden_dim)

        # Multi-layer LSTM with skip connections
        self.lstm_layers = nn.ModuleList([
            nn.LSTM(hidden_dim, hidden_dim, batch_first=True)
            for _ in range(num_layers)
        ])

        # Genre-specific layers
        self.genre_specific = nn.ModuleDict({
            'time_scale': nn.Linear(hidden_dim, 1),
            'note_density': nn.Linear(hidden_dim, 1),
            'velocity_scale': nn.Linear(hidden_dim, 1)
        })

        # Output processing
        self.output_layer = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim)
        )

        self.dropout = nn.Dropout(0.1)

    def forward(self, x, genre):

      batch_size = x.size(0)
      seq_length = x.size(1)

      # Genre parameters
      tempo_scale = torch.sigmoid(self.genre_tempo_base(genre))
      tempo_scale = tempo_scale.unsqueeze(-1).expand(-1, seq_length, -1)
      density_scale = torch.sigmoid(self.genre_density_scale(genre))
      density_scale = density_scale.unsqueeze(-1).expand(-1, seq_length, -1)

      # Pitch ranges
      pitch_low = self.pitch_range_low(genre)
      pitch_high = self.pitch_range_high(genre)

      # Convert bounds to tensors
      zeros = torch.zeros_like(pitch_low)
      max_low = torch.full_like(pitch_low, 88)
      max_high = torch.full_like(pitch_high, 108)

      # Clamp pitches
      pitch_low = torch.minimum(torch.maximum(pitch_low, zeros), max_low)
      pitch_high = torch.minimum(torch.maximum(pitch_high, pitch_low + 12), max_high)

      # Process through model
      x_embed = self.note_embedding(x)
      genre_embed = self.genre_embedding(genre).unsqueeze(1).expand(-1, seq_length, -1)
      pos_embed = self.position_embedding(
          self.position_ids[:, :seq_length]
      ).expand(batch_size, -1, -1)

      hidden = self.dropout(x_embed + genre_embed + pos_embed)

      lstm_outputs = []
      for lstm in self.lstm_layers:
          lstm_out, _ = lstm(hidden)
          lstm_outputs.append(lstm_out)
          hidden = self.dropout(hidden + lstm_out)

      combined = sum(lstm_outputs) / len(lstm_outputs)

      # Genre-specific scaling
      time_scale = torch.sigmoid(self.genre_specific['time_scale'](genre_embed))
      note_density = torch.sigmoid(self.genre_specific['note_density'](genre_embed))
      velocity_scale = torch.sigmoid(self.genre_specific['velocity_scale'](genre_embed))

      # Expand pitch ranges to match sequence dimensions
      pitch_low = pitch_low.view(batch_size, 1).expand(batch_size, seq_length)
      pitch_high = pitch_high.view(batch_size, 1).expand(batch_size, seq_length)

      # Output processing
      output = self.output_layer(combined)
      output_split = torch.split(output, 1, dim=-1)

      output = torch.cat([
          output_split[0] * time_scale,
          torch.clamp(output_split[1], pitch_low.unsqueeze(-1)/127.0, pitch_high.unsqueeze(-1)/127.0),
          output_split[2] * velocity_scale,
          output_split[3] * note_density
      ], dim=-1)

      return output
```

```python
def midi_to_sequence(midi):
    try:
        # Extract notes and convert to sequence format
        notes = []
        for note in midi.instruments[0].notes:
            notes.append([
                note.start,
                note.pitch / 127.0,  # Normalize pitch
                note.velocity / 127.0,  # Normalize velocity
                (note.end - note.start)  # Duration
            ])
        return np.array(notes)
    except:
        return None

class MidiDataset(Dataset):
    def __init__(self, midi_files, genre_labels, genre_to_index, max_seq_length=2048):
        self.data = []
        self.max_seq_length = max_seq_length

        for midi, genre in zip(midi_files, genre_labels):
            sequence = self.midi_to_sequence(midi)
            if sequence is not None:
                # Pad or truncate sequence to max_seq_length
                if sequence.size(0) > max_seq_length:
                    sequence = sequence[:max_seq_length]
                elif sequence.size(0) < max_seq_length:
                    padding = torch.zeros(max_seq_length - sequence.size(0), 4)
                    sequence = torch.cat([sequence, padding], dim=0)

                self.data.append((sequence, genre_to_index[genre]))

    def midi_to_sequence(self, midi):
        try:
            # Extract notes and convert to sequence format
            notes = []
            for note in midi.instruments[0].notes:
                notes.append([
                    note.start,
                    note.pitch / 127.0,  # Normalize pitch
                    note.velocity / 127.0,  # Normalize velocity
                    (note.end - note.start)  # Duration
                ])
            return torch.FloatTensor(notes)
        except:
            return None

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sequence, genre = self.data[idx]
        return sequence, genre
```

```python
def pitch_range_loss(output, target):
    # Extract pitch values
    output_pitch = output[:,:,1]
    target_pitch = target[:,:,1]

    # Calculate pitch ranges
    output_range = torch.max(output_pitch, dim=1)[0] - torch.min(output_pitch, dim=1)[0]
    target_range = torch.max(target_pitch, dim=1)[0] - torch.min(target_pitch, dim=1)[0]

    return F.mse_loss(output_range, target_range)



# Custom Loss Function
class EnhancedMusicLoss(nn.Module):
    def __init__(self):
        super(EnhancedMusicLoss, self).__init__()
        self.mse = nn.MSELoss(reduction='none')

    def forward(self, pred, target):
        # Basic reconstruction loss
        mse_loss = self.mse(pred, target)

        # Weight different note attributes
        weights = torch.tensor([1.0, 2.0, 0.5, 1.5]).to(pred.device)
        weighted_loss = mse_loss * weights.view(1, 1, -1)

        # Add pitch range loss
        range_loss = pitch_range_loss(pred, target)

        # Add note density loss - convert to tensor
        density_loss = torch.tensor(abs(pred.size(1) - target.size(1)),
                                  dtype=torch.float32,
                                  device=pred.device)

        return weighted_loss.mean() + range_loss * 0.5 + density_loss * 0.1
```

```python
def train_enhanced_model(model, dataloader, optimizer, criterion, epochs=100, device='cuda'):
    model.to(device)
    best_loss = float('inf')
    epoch_losses = []  # Track losses

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        batch_losses = 0  # Add counter for number of batches
        num_batches = len(dataloader)

        for batch_idx, (inputs, genres) in enumerate(dataloader):
            # Split long sequences into chunks
            chunk_size = model.max_sequence_length
            for i in range(0, inputs.size(1), chunk_size):
                chunk_inputs = inputs[:, i:i+chunk_size].to(device)
                if chunk_inputs.size(1) < 32:  # Skip very small chunks
                    continue

                optimizer.zero_grad()
                outputs = model(chunk_inputs, genres.to(device))
                loss = criterion(outputs, chunk_inputs)

                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()

                total_loss += loss.item()
                batch_losses += 1  # Increment counter

                if batch_idx % 10 == 0:
                    print(f'Epoch {epoch+1}/{epochs} [{batch_idx}/{num_batches}] '
                          f'Chunk {i//chunk_size} Loss: {loss.item():.4f}')

        # Calculate average loss correctly
        avg_loss = total_loss / batch_losses if batch_losses > 0 else total_loss
        epoch_losses.append(avg_loss)
        print(f'Epoch {epoch+1}/{epochs} Average Loss: {avg_loss:.4f}')

        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save(model.state_dict(), 'best_model.pth')

    # Only plot if we have more than one epoch
    if len(epoch_losses) > 1:
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(epoch_losses) + 1), epoch_losses, 'b-', label='Training Loss')
        plt.title('LSTM Model Training Progress')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.grid(True)
        plt.legend()
        plt.savefig('training_loss.png')
        plt.close()

    return epoch_losses  # Return the losses for potential later use
```

```python
def generate_enhanced_midi(model, input_midi, genre_index, temperature=1.0, device='cuda'):
    model.eval()

    # Convert input MIDI to sequence
    input_sequence = midi_to_sequence(input_midi)
    sequence_tensor = torch.FloatTensor(input_sequence)

    # Process in chunks with overlap for smoother transitions
    chunk_size = model.max_sequence_length
    overlap = chunk_size // 4
    outputs = []

    with torch.no_grad():
        for i in range(0, len(sequence_tensor), chunk_size - overlap):
            chunk = sequence_tensor[i:i + chunk_size]
            if len(chunk) < 32:
                continue

            chunk = chunk.unsqueeze(0).to(device)
            genre_tensor = torch.tensor([genre_index], device=device)

            output = model(chunk, genre_tensor)
            output = output.squeeze(0).cpu()

            # Enhanced variation with controlled randomness
            output = output / temperature
            noise = torch.randn_like(output) * 0.03 * (1 + temperature)
            output = output + noise

            # Smooth transitions between chunks
            if outputs and overlap > 0:
                prev_output = outputs[-1]
                blend_region = min(overlap, len(prev_output), len(output))
                weights = torch.linspace(0, 1, blend_region)
                output[:blend_region] = (
                    output[:blend_region] * weights.unsqueeze(1) +
                    prev_output[-blend_region:] * (1 - weights.unsqueeze(1))
                )

            outputs.append(output)

    # Combine chunks
    final_output = torch.cat(outputs, dim=0)

    # Normalize output
    final_output = torch.sigmoid(final_output)
    final_output = final_output.numpy()

    # Get genre name from index
    genre_name = index_to_genre[genre_index]

    return tensor_to_midi(final_output, genre_name)
```

```python
def analyze_genre_characteristics(midi_file):
    pm = pretty_midi.PrettyMIDI(midi_file)
    total_velocity = 0
    total_duration = 0
    pitches = []

    for instrument in pm.instruments:
        for note in instrument.notes:
            total_velocity += note.velocity
            total_duration += note.end - note.start
            pitches.append(note.pitch)

    note_count = len(pitches)
    if note_count > 0:
        avg_velocity = total_velocity / note_count
        avg_duration = total_duration / note_count
        pitch_range = max(pitches) - min(pitches) if pitches else 0

        return {
            'avg_velocity': avg_velocity,
            'avg_duration': avg_duration,
            'pitch_range': pitch_range,
            'density': note_count / pm.get_end_time() if pm.get_end_time() > 0 else 0
        }
    return None

# Analyze original and all generated versions
print("Original Analysis:")
orig_stats = analyze_genre_characteristics(original_file)
print(f"Avg Velocity: {orig_stats['avg_velocity']:.2f}")
print(f"Avg Duration: {orig_stats['avg_duration']:.2f}s")
print(f"Pitch Range: {orig_stats['pitch_range']}")
print(f"Note Density: {orig_stats['density']:.2f} notes/second\n")

for genre_file in glob.glob('/content/output_enhanced_*.mid'):
    genre = genre_file.split('_')[-1].replace('.mid', '')
    print(f"{genre} Analysis:")
    stats = analyze_genre_characteristics(genre_file)
    print(f"Avg Velocity: {stats['avg_velocity']:.2f}")
    print(f"Avg Duration: {stats['avg_duration']:.2f}s")
    print(f"Pitch Range: {stats['pitch_range']}")
    print(f"Note Density: {stats['density']:.2f} notes/second\n")
```

```python
import pretty_midi
import glob

def analyze_musical_features(midi_file, genre_name=""):
    pm = pretty_midi.PrettyMIDI(midi_file)

    # Get basic statistics
    tempo = pm.get_tempo_changes()[1].mean() if len(pm.get_tempo_changes()[1]) > 0 else 0
    pitch_range = [note.pitch for inst in pm.instruments for note in inst.notes]

    # Calculate note density (notes per second)
    duration = pm.get_end_time()
    total_notes = len(pitch_range)
    note_density = total_notes / duration if duration > 0 else 0

    print(f"\nAnalysis for {genre_name if genre_name else 'unknown genre'}:")
    print(f"Average tempo: {tempo:.2f} BPM")
    print(f"Pitch range: {min(pitch_range)} to {max(pitch_range)}")
    print(f"Unique pitches: {len(set(pitch_range))}")
    print(f"Note density: {note_density:.2f} notes/second")
    print("-" * 50)

    return {
        'tempo': tempo,
        'pitch_min': min(pitch_range),
        'pitch_max': max(pitch_range),
        'unique_pitches': len(set(pitch_range)),
        'note_density': note_density
    }

# Analyze original file
original_file = '/content/CS4710-Final-Project/lmd_matched_clean/Pop/TRAAAGR128F425B14B_dac3cdd0db6341d8dc14641e44ed0d44_Pop.mid'
print("\nOriginal File Analysis:")
original_stats = analyze_musical_features(original_file, "Original")

# Analyze all generated files
generated_files = glob.glob('/content/output_enhanced_*.mid')
genre_stats = {}

for gen_file in generated_files:
    genre = gen_file.split('_')[-1].replace('.mid', '')
    stats = analyze_musical_features(gen_file, genre)
    genre_stats[genre] = stats

# Compare key differences across genres
print("\nComparison across genres:")
print(f"{'Genre':<12} {'Tempo':<8} {'Note Density':<12} {'Pitch Range'}")
print("-" * 50)
for genre, stats in genre_stats.items():
    print(f"{genre:<12} {stats['tempo']:<8.1f} {stats['note_density']:<12.2f} {stats['pitch_min']}-{stats['pitch_max']}")
```

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile

def plot_waveforms(file_paths, duration=5):
    """
    Plots the waveforms for the first few seconds of multiple .wav files with titles.

    Parameters:
        file_paths (list): List of file paths to .wav files.
        duration (int): Duration in seconds to plot for each file (default: 5 seconds).
    """
    for file_path in file_paths:
        try:
            # Read the .wav file
            sample_rate, data = wavfile.read(file_path)

            # Calculate the number of samples for the specified duration
            num_samples = int(sample_rate * duration)

            # Ensure the data length supports slicing
            if len(data) < num_samples:
                print(f"Warning: File {file_path} is shorter than {duration} seconds.")
                num_samples = len(data)

            # Slice the data for the specified duration
            data_sliced = data[:num_samples]
            time = [i / sample_rate for i in range(len(data_sliced))]

            # Generate a title for the graph
            title = f'Waveform: {file_path.split("/")[-1]} (First {duration} Seconds)'

            # Plot the waveform
            plt.figure(figsize=(12, 6))
            plt.plot(time, data_sliced)
            plt.title(title)
            plt.xlabel('Time (seconds)')
            plt.ylabel('Amplitude')
            plt.grid(True)
            plt.show()

        except Exception as e:
            print(f"Error processing file {file_path}: {e}")
```

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine

def load_audio(filepath):
    """Load a .wav file and return the waveform and sampling rate."""
    try:
        audio, sr = librosa.load(filepath, sr=None, mono=True)  # Load in mono for consistency
        return audio, sr
    except Exception as e:
        raise ValueError(f"Error loading audio file {filepath}: {e}")

def plot_waveform(audio, sr, title):
    """Plot the waveform of an audio signal."""
    plt.figure(figsize=(10, 4))
    librosa.display.waveshow(audio, sr=sr)
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()

def plot_spectrogram(audio, sr, title):
    """Plot the spectrogram of an audio signal."""
    plt.figure(figsize=(10, 4))
    S = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=128, fmax=8000)
    S_dB = librosa.power_to_db(S, ref=np.max)
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8000, cmap='coolwarm')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()

def compute_mfcc(audio, sr, n_mfcc=13):
    """Compute MFCC features for an audio signal."""
    return librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)

def plot_mfcc_similarity(audio1, sr1, audio2, sr2):
    """Compute and plot the similarity of MFCCs."""
    mfcc1 = compute_mfcc(audio1, sr1)
    mfcc2 = compute_mfcc(audio2, sr2)

    # Plot MFCCs
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    librosa.display.specshow(mfcc1, sr=sr1, x_axis='time', cmap='coolwarm')
    plt.colorbar()
    plt.title("MFCCs of File 1")

    plt.subplot(1, 2, 2)
    librosa.display.specshow(mfcc2, sr=sr2, x_axis='time', cmap='coolwarm')
    plt.colorbar()
    plt.title("MFCCs of File 2")

    plt.tight_layout()
    plt.show()

    # Compute cosine similarity
    mfcc1_mean = mfcc1.mean(axis=1)
    mfcc2_mean = mfcc2.mean(axis=1)
    similarity = 1 - cosine(mfcc1_mean, mfcc2_mean)

    print(f"Spectral similarity (MFCC-based): {similarity:.2f}")
```
