```python
import json
import os
import pretty_midi
import glob

base_directory = '.\lmd_matched_clean'
# Specify the path to your JSON file
file_path = '.\cleaned_data.json'

# Open and read the JSON file
with open(file_path, 'r') as file:
    base_info = json.load(file)

# Iterate over each entry in the JSON data
for entry in base_info[:100]:
    # Extract the seed_genre and track_id
    seed_genre = entry['seed_genre']
    track_id = entry['track_id']

    # Construct the MIDI file path
    midi_dir = os.path.join(base_directory, seed_genre)
    pattern = os.path.join(midi_dir, f"{track_id}*.mid")
    matching_files = glob.glob(pattern)

    if matching_files:
        # Load the first matching file
        midi_file_path = matching_files[0]
        try:
            # Load the MIDI file using pretty_midi
            midi = pretty_midi.PrettyMIDI(midi_file_path)
            print(f"Successfully loaded: {midi_file_path}")
            for instrument in midi.instruments:
                print(f"Instrument: {instrument.name}, Program: {instrument.program}, Is Drum: {instrument.is_drum}")
        except Exception as e:
            print(f"Error loading {midi_file_path}: {e}")
    else:
        print(f"No matching files found for track_id: {track_id} in {midi_dir}")
```
